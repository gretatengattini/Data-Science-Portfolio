```python
# IMPORTING LIBRARIES
import pandas as pd                   # DATA SCIENCE ESSENTIALS
import matplotlib.pyplot as plt       # ESSENTIAL GRAPHIC OUTPUT
import seaborn as sns                 # ENHANCED GRAPHIC OUTPUT
import numpy as np                    # MATHEMATICALS ESSENTIALS
import statsmodels.formula.api as smf # REGRESSION MODELLING


# SETTING PANDAS PRINT OPTIONS
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


# SPECIFYING FILE PATHS
file = './datasets/Ames_Housing_Dataset.xlsx'


# READING THE FILE INTO PYTHON
housing = pd.read_excel(io     = file,
                        header = 0   )

# DISPLAYING DATASET COLUMNS FOR VISUALIZATION 
housing.info()
```

```python
# SETTING 'log_Sale_Price' FOR NORMALIZATION
housing['log_Sale_Price'] = np.log1p(housing['Sale_Price'])

# REMOVING NON-NUMERIC VARIABLES FROM THE DATASET
housing = housing.drop(["Street", "Lot_Config", "Neighborhood"], axis = 1)

# CREATING A PEARSON-BASED CORRELATION MATRIX
df_corr = housing.corr().round(2)

# PRINTING PEARSON-BASED CORRELATIONS WITH SALE PRICE INFORMATION TO CHECK NORMALIZATION
# AND DATA CHANGES BASED ON OUTLIERS REMOVAL
df_corr.loc[ :, ['Sale_Price', 'log_Sale_Price'] ].sort_values(by = 'Sale_Price',
                                                                ascending = False)
```

<hr style="height:.9px;border:none;color:#333;background-color:#333;" />
<h3>Step 2: Code the New Features</h3><br>
Use the code cell below to develop the features you have hypothesized.

```python
## FEATURE 1 - 'total_added_space' ##

# 'total_added_space' IS A FEATURE THAT SUMS THE SQUARED FOOTAGE OF THE PORCH AND GARAGE AREAS

housing['total_added_space'] = housing['Porch_Area'] + housing['Garage_Area'] # ADDING FORMULA AND CREATING NEW FEATURE
housing.head() # CHECKING RESULTS
```

<br>

```python
## FEATURE 2 - 'baths_rooms_ratio' ##

# 'baths_rooms_ratio' IS A FEATURE THAT CALCULATES THE RATIO OF BATHROOMS PER ROOM IN A PROPERTY

housing['baths_rooms_ratio'] = (housing['Full_Bath'] / housing['TotRms_AbvGr']).round(2)
# ADDING FORMULA, CREATING NEW FEATURE AND ROUNDING IT TO 2 DECIMAL POINTS
housing['baths_rooms_ratio'].head() # CHECKING RESULTS
```

<br>

```python
## FEATURE 3 - '2suv_accommodation' ##

# '2suv_accommodation' IS A FEATURE WITH BOOLEAN LOGIC THAT INDICATES IF A GARAGE HAS ENOUGH SPACE TO ACCOMMODATE 2 LARGE SUVs
housing['2suv_accommodation'] = 0 # CREATING COLUMN WITH 0 AS PLACEHOLDERS

for index, row in housing.iterrows(): # CREATING DATA IMPUTATION LOGIC
    if housing.loc[index, 'Garage_Cars'] > 1 and housing.loc[index, 'Garage_Area'] > 400 : # SETTING CONDITIONS BASED ON RESEARCH
            housing.loc[index, '2suv_accommodation'] = 1 # DATA IMPUTATION LOGIC IF TRUE
    else :
        housing.loc[index, '2suv_accommodation'] = 0 # DATA IMPUTATION LOGIC IF FALSE

housing.head() # CHECKING RESULTS
```

<hr style="height:.9px;border:none;color:#333;background-color:#333;" />
<h3>Step 3: Check the Results</h3><br>
Develop a correlation matrix or a heatmap to show the linear relationships between your five new features and the response variables (<em>Sale_Price</em> AND <em>log_Sale_Price</em>).

```python
# CODE HERE

features = ['total_added_space', 'baths_rooms_ratio', '2suv_accommodation', 'Sale_Price', 'log_Sale_Price'] # LISTING CREATED FEATURES AND SALE PRICE

corr_m = housing[features].corr() # DEFINING CORRELATION METHOD

corr_m.loc[['total_added_space', 'baths_rooms_ratio', '2suv_accommodation'], # RUNNING CORRELATION METHOD
           ['Sale_Price', 'log_Sale_Price']].sort_values(by = 'Sale_Price', # ASCENDING RESULTS BY 'Sale_Price'
                                                        ascending = True) 
```
