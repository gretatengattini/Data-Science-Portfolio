## Finding Nemo Adventure Game

"Finding Nemo" is a 2003 animated film produced by Pixar Animation Studios for Walt Disney Pictures (Stanton, A., & Unkrich, L. (Directors). (2003). _Finding Nemo_ [Film]. Pixar Animation Studios; Walt Disney Pictures).

Directed by Andrew Stanton and Lee Unkrich, the movie follows Marlin, an overprotective clownfish, whose son Nemo is captured by scuba divers and ends up in a fish tank in a dentist's office in Syndey. Marlin, accompanied by Dory, a forgetful blue fish, embarks on a journey to rescue Nemo. Meanwhile, Nemo, determined to escape the dentistâ€™s tank, collaborates with the Tank Gang, a group of aquarium fish. Marlin and Dory face various challenges, including encounters with sharks, jellyfish, and a whale. Ultimately, Nemo successfully blocks the tank's filter, causing chaos in the dentist's office. The story converges as Marlin, Dory, and Nemo reunite and, with the help of other fish, escape a fishing trawler's net. The film ends with Marlin more confident, Nemo heading to school, and the Tank Gang escaping into Sydney Harbor.

In this Python game, designed based on the story of Finding Nemo, the objective is to reunite Nemo with his dad. The player navigates through 4 stages on an adventure with Marlin, Dory, and Nemo. Each stage is unique, with elements of randomness such that the game can be played at least three times and be completely different each time. The game includes puzzles and riddles, each with a varying nature (e.g. navigate a map, guess a number, solve a word riddle, etc.).

```python
import random

def stage1():

    print(f"""
WELCOME TO FINDING NEMO

You're in the Great Barrier Reef for Nemo's first day of school. ðŸŽ’
While Marlin talks to the teacher, Nemo gets captured by a scuba diver! ðŸ˜°
The man loses his mask ðŸ¥½, which has the destination address.
Then, Marlin meets Dory, a fish with a bad memory but can read. ðŸ¤“
Help Dory figure out the address! She is having trouble reading the number...
""")

    # The correct number is defined randomly
    correct_address_num = random.randint(a=1, b=10)
    remaining_guesses = 3


    while remaining_guesses > 0:

        # The player will pick a random number from 1 to 10
        address_guess = input(prompt='What is the address number on the ' +
                              'mask, between 1 and 10?\n> ')

        # Convert input to integer, for numerical comparison
        try:
            int_guess = int(address_guess)
        except:
            print('Invalid input. Please enter a number between 1 and 10.')
            continue

        if int_guess < 1 or int_guess > 10:
            print('Your guess is out of range, please enter a number ' +
                  'between 1 and 10.')

        # Check if the guess is correct
        elif int_guess == correct_address_num:
            print(f'You got it!ðŸ˜¯ ðŸŽ‰ Nemo was taken to {correct_address_num} ' +
                  'Wallaby Way, Sydney ðŸ“')
            break

        # If the guess is incorrect
        else:
            remaining_guesses -= 1

            if remaining_guesses == 0:
                print('You are out of guesses ðŸ™. The address is ' +
                      f'{correct_address_num} Wallaby Way, Sydney.')

            # If the player has more guesses, receive a hint
            else:
                if int_guess > correct_address_num:
                    print(f'Too high! Try again. {remaining_guesses} guesses left')
                else:
                    print(f'Too low! Try again. {remaining_guesses} guesses left')

    # Pause for the player to read the end of the stage, and continue the story.
    input(prompt='Press return to continue the adventure!\n> ')

def stage2():
    print("""
Suddenly, Marlin and Dory see a school of sunfish that exclaim, "For Sydney,
you have to go this way â¬…ï¸ â¬†ï¸ â¤µï¸ !". Marlin and Dory end up in a maze
of jellyfish. ðŸ™ˆ Help them escape the jellyfish forest! ðŸ™ðŸ»
""")

    # We give just 2 guesses here so that the game is faster (within the 3min
    # time limit)
    remaining_guesses = 2
    correct_guesses = 0

    # There are different types of jellyfish that will appear in front of them
    possible_jellyfish_types = ['angry', 'tiny', 'upside down', 'green', 'lazy']

    while remaining_guesses > 0:

        # Depending on the jellyfish type, they have to determine the right
        # move
        current_jellyfish = random.choice(possible_jellyfish_types)
        direction_guess = input(prompt=f'A {current_jellyfish} jellyfish is in front ' +
                                'of them! Should they go left or right?\n> ')
        # Add an extra space for better UI
        print('')

        if current_jellyfish == 'angry':
            print(f'Oh no! They try to go {direction_guess}, but the angry jellyfish ' +
                  'stings them ðŸ˜«')

        elif current_jellyfish == 'tiny':
            print(f'Since the jellyfish is so small, they go {direction_guess} and ' +
                  'easily get past it!')
            correct_guesses += 1

        elif current_jellyfish == 'upside down':
            if direction_guess == 'right':
                print('Good job! They go right and swim around the ' +
                      'upside-down jellyfish. ðŸ™Œ')
                correct_guesses += 1
            else:
                print(f'Oh no! They try to go {direction_guess}, but the upside-down ' +
                      'jellyfish stings them âš¡ï¸')

        elif current_jellyfish == 'green':
            if direction_guess == 'left':
                print('Good job! They go left and swim around the green ' +
                      'jellyfish. ðŸ™Œ')
                correct_guesses += 1
            else:
                print(f'Oh no! They try to go {direction_guess}, but the green ' +
                      'jellyfish stings them âš¡ï¸')

        elif current_jellyfish == 'lazy':
            print(f'Since the jellyfish is lazy, they go {direction_guess} and it ' +
                  'is too lazy to sting them -- they get past it!')
            correct_guesses += 1
        else:
            # Note that because the cases are exhaustive above, we should
            # never run this line
            print('')

        remaining_guesses -= 1

        # If they guess correctly
        if correct_guesses == 2:
            print('Woo. Amazing job! ðŸŽ‰ Dory and Marlin escaped the ' +
                  'jellyfish and are free to continue their journey!')
            break
        elif remaining_guesses == 0:
            print(
'''Thank goodness, the jellyfish maze clears.
Dory and Marlin recover and continue their journey.
''')
        else:
            print('The jellyfish maze continues...')

    # Pause for the player to read the end of the stage, and continue the story.
    input(prompt='Press return to continue the adventure!\n> ')

def stage3():
    print("""
Nemo is in a Sydney dentist's aquarium ðŸ¦· . Meanwhile,
Dory and Marlin woke up in the Australian current, on the backs of a group
of sea turtles ðŸ¢. Nemo hears of their journey from a pelican and
gains courage to escape the tank.
""")

    # Randomly determine the type of question; there are 3 different tasks
    challenge_type = random.choice(['sequence', 'avoid', 'complete'])

    if challenge_type == 'sequence':
        # Sequence-based question - there are 3 moves that need to be ordered
        # by the player
        sequence = [
            '1. Wait for the dentist to change the filter, and remove the fish from the aquarium.',
            '2. Put a pebble in the filter, so that the water gets dirty.',
            '3. Jump into the toilet and swim into the bay.'
        ]

        print(
'''Nemo decides to make an escape plan. Help him figure out the order of
actions to take:''')
        for action in sequence:
            print('  ' + action)

        remaining_guesses = 2

        while remaining_guesses > 0:
            # Ask the player to choose the correct sequence of moves
            choice = input(prompt=('Enter the correct sequence of moves ' +
                           'as numbers with single spaces (e.g., 1 2 3). ' +
                           f'You have {remaining_guesses} chances:\n> '))

            # Remove any extra spaces, to improve user experience
            choice = choice.strip()

            # Check if the player's choice matches the correct sequence
            if choice == '2 1 3':
                print('Congratulations! Nemo is ready to escape the aquarium.')
                break

            # Otherwise, the guess is incorrect
            else:
                remaining_guesses -= 1
                if remaining_guesses > 0:
                    print(f'Incorrect sequence. You have {remaining_guesses} ' +
                          'chances left.')

                    # Now, we give some clues, depending on the player guesses
                    if choice[-1] != '3':
                        print('Hint: the final thing Nemo should do is ' +
                              'swim into the bay.')
                    elif choice[0] == '3':
                        print("It's a bit strange for Nemo to go inside " +
                              "the ðŸš½ before the water gets dirty in the tank.")
                    # Although it's not necessary, put an else here because
                    # it's a requirement.
                    else:
                        print('Try again')
                else:
                    print(f'You have run out of chances. ðŸ™ ' +
                          'The correct order was: 2 1 3.')

    elif challenge_type == 'avoid':
        # Avoid-based question - Multiple choice question
        remaining_guesses = 2

        print(
"""In order to escape, what should Nemo avoid doing? Choose one of the
following options:
    1. Stay focused on the plan
    2. Demoralize himself just before carrying out his plan
    3. Sleep a lot the day before""")

        correct_answer = '2'

        while remaining_guesses > 0:
            # Ask the player what Nemo should avoid
            choice = input(prompt='> ')

            # Remove extra spaces and lowercase input, for better user experience
            choice = choice.strip().lower()

            # Check if the player's choice matches the correct answer
            if (choice == correct_answer or
                choice == 'demoralize himself just before carrying out his plan'):
                print('Congratulations! Nemo knows what to avoid. ðŸ‘ðŸ» He ' +
                      'is ready to escape the aquarium. ðŸŽ‰')
                break

            # The guess is incorrect
            else:
                remaining_guesses -= 1

                if remaining_guesses > 0:
                    if choice == '3' or choice == 'sleep a lot the day before':
                        print(f'Incorrect. There is nothing wrong with ' +
                              'sleeping. You have {guesses} chances left.')
                    elif choice == '1' or choice == 'stay focused on the plan':
                        print('Incorrect. That is the most important ' +
                              f'thing he should do. You have {guesses} ' +
                              'chances left.')
                    else:
                        print('Incorrect. Try again, the answer is not ' +
                              f'too difficult. You have {guesses} ' +
                              'chances left.')
                else:
                    print('You have run out of chances. The correct ' +
                          'answer was: {correct_answer}.')

    else:
        # Ask the player to complete the sentence question
        sentence = "Pretend to be...?"
        correct_answer = 'dead'
        remaining_guesses = 2

        print("What should Nemo do to avoid getting taken by the " +
              "dentist's nephew? He can pretend to do something...")

        while remaining_guesses > 0:
            # Ask the player to complete the sentence
            choice = input(prompt=f"Complete the sentence: {sentence}\n> ")

            # Remove extra spaces and lowercase, for better user experience
            choice = choice.strip().lower()

            # Check if the player's choice matches the correct answer
            if choice == correct_answer:
                print("Congratulations! ðŸŽ‰ That's correct! Nemo is safe!")
                break

            # The guess is incorrect
            else:
                remaining_guesses -= 1

                if remaining_guesses == 0:
                    print("You have run out of chances. ðŸ™ The correct " +
                          f"answer was: {correct_answer}.")
                elif remaining_guesses == 1:
                    print("Incorrect. Hint: It's a single word, " +
                          f"with 4ï¸âƒ£ letters. You have {remaining_guesses} " +
                          "chances left. ðŸ’ªðŸ¼")
                else:
                    print("Incorrect. Hint: it's something dramatic... " +
                          f"You have {remaining_guesses} chances left.")

    # Pause for the player to read the end of the stage, and continue the story.
    input(prompt='Press return to continue the adventure!\n> ')

def win():
    # If the player wins a congratulation message appears
    print(
'''ðŸŽ‰ Congratulations! ðŸŽ‰ You WIN!

Nemo and his father have been reunited because of you. ðŸ’™''')

def fail():
    # If the player loses, they will need to try again.
    print('Oh no! You lost. ðŸ˜” Try again to reunite Nemo and his Father. ðŸ™ðŸ»')

def stage4():
    print('''
Nemo is able to escape and meets Dory, but she doesn't remember anything!
Dory needs her memory back so she can reunite Nemo and his father.
You have just one chance!''')

    # Randomly select one of the 3 games to play
    game_id = random.randint(a=1, b=3)
    player_wins = False

    # In each game, update the player_wins variable with the outcome
    if game_id == 1:
        choice = input(prompt="""
Which of these words is the anagram of a piece of information to refresh Dory's
memory? Choose a number between 1 and 3:
    1. Disney
    2. Pixar
    3. DreamWork
> """)
        if choice.strip() == '1':
            print("That's right! Disney is an anagram of Sidney. Dory remembers everything!")
            player_wins = True
        else:
            player_wins = False

    elif game_id == 2:
        choice = input(prompt="""
Which question should Dory ask Nemo? The answer to this question should
bring back her memory. Choose either '1' or '2':
    1. How many siblings do you have?
    2. What are your parents' names?
> """)

        if choice.strip() == '2':
            print("That's right! Nemo says Marlin, and Dory remembers everything.")
            player_wins = True
        else:
            player_wins = False

    else:
        choice = input(prompt="""
Nemo helps Dory remember, but then suddenly, she gets stuck in a huge fishing net! ðŸ˜§

What should Nemo advise her to do to escape? Choose 1, 2, or 3:
    1. Swim with the direction the net is moving
    2. Swim against the direction of the net
    3. Stay as still as possible
> """)

        if choice.strip() == '2':
            print("That's right! She breaks free and reunites Nemo and Marlin.")
            player_wins = True
        else:
            player_wins = False

    # Finally, depending on the game outcome, the player either wins or fails
    if player_wins:
        win()
    else:
        fail()

# The main game function
def play_game():
    print(f"""
{'-' * 78}

DIRECTIONS:

Welcome to Finding Nemo! The game has 4 stages, each with one task. You'll
help a fish, Marlin, and his son, Nemo, go through an adventure to be
reunited!

The last stage will determine whether you win. Let's go!

{ '-' * 78}\n""")

    # The player plays through each stage, one at a time.
    input(prompt="Press Enter to continue.\n> ")
    stage1()
    stage2()
    stage3()
    stage4()
    print('Thank you for playing!')

# The function call that starts the game
play_game()
```
